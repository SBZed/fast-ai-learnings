````python
import os
from getpass import getpass
import urllib

username = 'SBZed'
repo = 'fast-ai-learnings'
password = getpass('Password: ')
password = urllib.parse.quote(password) # your password is converted into url format

cmd_string = f'git remote add origin https://{username}:{password}@github.com/{username}/{repo}.git'

os.system(cmd_string)
cmd_string, password = "", "" # removing the password from the variable
````

# 2. Hide Password



- To avoid password exposure we are using `getpass` library. so that, we don't push code containing our sensitive information.
- Your password is only get store in `config` file of your `.git` directory of local repository (in this case, our google drive folder). e.g.
```json
// config
[core]
	repositoryformatversion = 0
	filemode = true
	bare = false
	logallrefupdates = true
[remote "origin"]
	url = https://SBZed:{my_password}@github.com/SBZed/fast-ai-learnings.git
	fetch = +refs/heads/*:refs/remotes/origin/*
```

- `.git` directory is not get push to our remote repository/github while doing commit.
- When someone clone your repository they get public url. e.g.
```
https://github.com/SBZed/fast-ai-learnings.git
```
